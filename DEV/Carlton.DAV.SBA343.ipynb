```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Section 1
# Task 1: Read data from CSV file
print("Task 1: Reading data from CSV file")
df = pd.read_csv(R"C:\Users\carlton.powell\Downloads\woocommerce-product-export.csv")

# Task 2: summary of columns
print("\nTask 2: Summary of columns using info() method")
df.info()

# Task 3: summary statistics
print("\nTask 3: Summary statistics")
print(df.describe())

# Task 4:first five rows
print("\nTask 4: First five rows")
print(df.head())

# Task 5:last five rows
print("\nTask 5: Last five rows")
print(df.tail())

# Task 6:total_profit and month_number columns
print("\nTask 6: total_profit and month_number columns")
print(df[['total_profit', 'month_number']])

# Task 7: total profit per month
print("\nTask 7: Bar plot of total profit per month")
plt.figure(figsize=(10, 6))
plt.bar(df['month_number'], df['total_profit'])
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Company Profit per Month')
plt.xticks(df['month_number'])
plt.savefig('bar_plot.png')
plt.show()

# Task 8: total profit
print("\nTask 8: Line plot of total profit")
plt.figure(figsize=(10, 6))
plt.plot(df['month_number'], df['total_profit'],
         linestyle=':', color='red', linewidth=3,
         marker='o', markerfacecolor='black')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Company Profit per Month')
plt.xticks(df['month_number'])
plt.legend(['Profit'], loc='lower right')
plt.savefig('line_plot.png')
plt.show()

# Task 9: Multi-line plot for all products
print("\nTask 9: Multi-line plot for all products")
# Get product columns (excluding non-product columns)
product_columns = df.columns[2:-1]  # Assuming first two and last columns are not products

plt.figure(figsize=(12, 6))
for product in product_columns:
    plt.plot(df['month_number'], df[product], label=product)

plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Product Sales Data')
plt.xticks(df['month_number'])
plt.legend(loc='upper left')
plt.savefig('multi_line_plot.png')
plt.show()

print("\nTask 10: Scatter plot for bathingsoap")
plt.figure(figsize=(10, 6))
plt.scatter(df['month_number'], df['bathingsoap'], s=100)
plt.grid(linestyle='--')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Bathingsoap Sales Data')
plt.xticks(df['month_number'])
plt.savefig('scatter_plot.png')
plt.show()

###date vs temperature
print("\nSection Two - Task 1: Line chart of date vs temperature")
date = ["25/12", "26/12", "27/12"]
temp = [8.5, 10.5, 6.8]

plt.figure(figsize=(8, 5))
plt.plot(date, temp, marker='o')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('Date-wise Temperature')
plt.savefig('temperature_plot.png')
plt.show()

# height vs weight
print("\nSection Two - Task 2: Line chart of height vs weight")
height = [121.9, 124.5, 129.5, 134.6, 139.7, 147.3, 152.4, 157.5, 162.6]
weight = [19.7, 21.3, 23.5, 25.9, 28.5, 32.1, 35.7, 39.6, 43.2]

plt.figure(figsize=(10, 6))
plt.plot(weight, height, linestyle='-.', color='green',
         marker='o', markersize=10, markerfacecolor='green')
plt.xlabel('Weight in kg')
plt.ylabel('Height in cm')
plt.title('Average Weight with Respect to Average Height')
plt.legend(['Height-Weight'], loc='lower right')
plt.savefig('height_weight_plot.png')
plt.show()
```
